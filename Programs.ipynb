###### Stacks & Queues 

```python
# Program for stack implementation

class Stack():
    """ Stack class """
    def __init__(self):
        """ Initializes stack class """
        self.n = 0
        self.A = []
        
    def push(self, value):
        """ Pushes element into the stack """
        self.A.append(value)
        self.n += 1
    
    def pop(self):
        """ Removes top element from stack and returns it """
        if self.n < 1:
            return "No elements are present in the stack"
        else:
            top_value = self.A[self.n-1]
            del self.A[self.n-1]
            self.n -= 1
            return top_value
    
    def peak(self):
            """ Returns top element """
            if self.n < 1:
                return "No elements are present in the stack"
            else:
                return self.A[self.n-1]
    def isEmpty(self):
        if self.n < 1:
            return True
        return False
    def size(self):
        return self.n

# Abstract Data Type Queue Implementation

class Queue():
    """ Abstract Data Type Queue Class"""
    def __init__(self):
        """ Initializes Queue """
        self.items = []
        
    def isEmpty(self):
        """ Returns True if queue is empty, else False """
        return self.items == []
    
    def enqueue(self, item):
        """ Insert value at the beginning """
        self.items.insert(0,item)
    
    def dequeue(self):
        """ Removes first inserted element """
        if len(self.items) != 0:
            return self.items.pop()
        else:
            return "No items found in the queue"
    
    def size(self):
        """ Returns the size of queue """
        return len(self.items)
```

###### Balanced parantheses 

```python
# Balanced Parantheses 

# Method-I

def is_balanced(string):
    d = {'[':']','{':'}','(':')'}
    if string[0] not in d.keys():
        return False
    else:
        q = []
        for i in string:
            if i in d.keys():
                q.append(i)
            else:
                if i == d[q[-1]]:
                    q.pop()
                else:
                    return False
    return True

print(is_balanced("{{{}}}"))


# Method-II

def balanced(string):
    opened = "{[("
    parantheses = [('{','}'),('[',']'),('(',')')]
    last_opened = []
    for i in string:
        if i in opened:
            last_opened.append(i)
        else:
            if (last_opened,i) in parantheses:
                last_opened.pop()
            else:
                return False
    return True

for i in range(5):
    string = input("Enter String: ")
    print(balanced(string))
```
